```python
import random
import matplotlib.pyplot as plt
import math
def activation(x):
  if x>0:
    return x
  else:
    return x*0.1

def sigmoid(x):
  return 1 / (1 + math.exp(-x))

w1 = [random.random()]
w2 = [random.random()]
w3 = [random.random()]
b1 = [random.random()]
b2 = [random.random()]
b3 = [random.random()]
def predicting(x):
  return abs(math.atan(x))/2
def calc(x):
  return sigmoid(activation(activation(activation(w1[0]*x+b1[0]))*w2[0]+b2[0])*w3[0]+b3[0])
lr = 0.01
h = 0.01
data = {}
lis = []
for i in range(0,1000):
    lis.append(predicting(i/1000))
li2s =[]
li = [w1,w2,w3,b1,b2,b3]
list_of_errors = []
for i in range(0,10000):
  a = random.uniform(0,1)
  data[a] = predicting(a)
for epoch in range(0,2000):
  lr = lr*0.9955
  if epoch % 100 == 0:
    print(epoch)
    print(lr)

    li2 = []
    for i in range(-1000,1000):
      li2.append(calc(i/1000))
    li2s.append(li2)
  real_error = []
  for key in data.keys():
    org_answer = calc(key)
    target = data[key]
    error_li = []

    error = abs(target-org_answer)
    for variable in li:
      variable[0]+=h
      new_ans = calc(key)
      new_error = abs(target-new_ans)
      variable[0]-=h
      error_li.append(error-new_error)
      real_error.append(new_error)
    #print(real_error)
    for i in range(0,len(error_li)):
      li[i][0]+=lr*error_li[i]
  list_of_errors.append(sum(real_error))
plt.plot(list_of_errors)
```

```python

for i in range(0,len(li2s)):
  plt.plot(lis)
  plt.plot(li2s[i])
  plt.show()
```
